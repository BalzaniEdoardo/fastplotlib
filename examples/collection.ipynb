```python
from fastplotlib import Plot
from ipywidgets import VBox, HBox, IntSlider
import numpy as np
```

```python
# linspace, create 100 evenly spaced x values from -10 to 10
xs = np.linspace(-10, 50, 100)
# sine wave
ys = np.sin(xs)
sine = np.dstack([xs, ys])[0]

# cosine wave
ys = np.cos(xs)
cosine = np.dstack([xs, ys])[0]
```

```python
coll = list()
for i in range(0, 10, 1):
    s = np.copy(sine)
    s[:, 1] += i * 5
    coll.append(s)
    
    c = np.copy(cosine)
    c[:, 1] += (i * 5) + 3
    coll.append(c)
```

```python
plot = Plot()
```

```python
lines = plot.add_line_collection(coll, cmap="jet")
```

```python
plot.show()
```

```python
lines[:].present.add_event_handler(plot.auto_scale)
```
